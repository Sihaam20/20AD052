```python
from crewai import Agent, Task, Crew
```

```python
import os
from crewai import LLM

gemini_api_key = os.getenv("GEMINI_API_KEY")

gemini_llm = LLM(
    model='gemini/gemini-3-flash-preview',
    api_key=gemini_api_key,
    temperature=1.0
)
```

```python
from crewai.tools import BaseTool

class AgenticResearchTool(BaseTool):
    name: str = "Agentic AI Research Tool"
    description: str = "Fetches latest trends and developments in agentic AI."

    def _run(self, topic: str) -> str:
        print(f"--- Researching topic: {topic} ---")
        return (
            """Latest Developments in Agentic AI:
- Rise of autonomous AI agents using frameworks like CrewAI and AutoGen.
- Increasing adoption of multi-agent workflows in enterprise automation.
- Integration of RAG (Retrieval-Augmented Generation) with agents.
- Growth of open-source ecosystems for agent orchestration.
- Focus on memory, planning, and tool usage in LLM agents.
- Emerging use cases in coding, research, and customer support automation."""
        )

research_tool = AgenticResearchTool()
```

```python
from crewai import Agent

# Agent 1: Researcher
researcher = Agent(
    role='Agentic AI Research Specialist',
    goal='Find and analyze the latest developments in agentic AI.',
    backstory=(
        """You are an AI researcher specializing in agentic systems and multi-agent workflows.
        You track cutting-edge advancements and summarize key trends."""
    ),
    verbose=True,
    allow_delegation=False,
    tools=[research_tool],
    llm=gemini_llm
)

# Agent 2: Writer
writer = Agent(
    role='Technical Blog Writer',
    goal='Write a detailed, engaging blog post on agentic AI.',
    backstory=(
        """You are a skilled technical writer who explains complex AI topics clearly.
        You create engaging, structured, and easy-to-read blog posts."""
    ),
    verbose=True,
    allow_delegation=False,
    llm=gemini_llm
)

# Agent 3: Editor
editor = Agent(
    role='Content Editor',
    goal='Refine and polish the blog content.',
    backstory=(
        """You are a professional editor ensuring clarity, flow, and quality.
        You make content concise, polished, and publication-ready."""
    ),
    verbose=True,
    allow_delegation=False,
    llm=gemini_llm
)
```

```python
from crewai import Task

# Task 1: Research
research_task = Task(
    description=(
        """Research the latest developments in agentic AI.
        Identify key trends, tools, frameworks, and real-world applications.
        Use the Agentic AI Research Tool."""
    ),
    expected_output=(
        """A structured summary including:
- Key trends in agentic AI
- Popular frameworks and tools
- Real-world use cases
- Future outlook"""
    ),
    agent=researcher
)

# Task 2: Write Blog
writing_task = Task(
    description=(
        """Using the research insights, write a detailed blog post titled:
        'Trying AI Agent Frameworks – CrewAI'.

        Include:
- Introduction to CrewAI
- Explanation of multi-agent workflows
- Use case: automated blog generation
- Step-by-step explanation
- Personal insights and reflections
- Conclusion"""
    ),
    expected_output=(
        """A well-structured blog post with headings, clear explanations,
        and engaging narrative suitable for developers."""
    ),
    agent=writer
)

# Task 3: Edit Blog
editing_task = Task(
    description=(
        """Refine the blog post for clarity, grammar, and readability.
        Ensure it is concise, professional, and engaging.
        Add proper formatting and improve flow."""
    ),
    expected_output=(
        """A polished, publication-ready blog post with:
- Clear structure
- Improved readability
- Professional tone
- No grammatical errors"""
    ),
    agent=editor
)
```

```python
from crewai import Crew, Process

blog_crew = Crew(
    agents=[researcher, writer, editor],
    tasks=[research_task, writing_task, editing_task],
    process=Process.sequential,
    verbose=True
)
```

```python
import time
from crewai import LLM

def create_gemini_llm():
    return LLM(
        model='gemini/gemini-3-flash-preview',
        api_key=os.getenv("GEMINI_API_KEY"),
        temperature=1.0
    )

def get_llm_with_retry(retries=3, delay=5):
    for i in range(retries):
        try:
            return create_gemini_llm()
        except Exception as e:
            print(f"Retry {i+1}/{retries} failed: {e}")
            time.sleep(delay)
    raise Exception("LLM failed after retries")

gemini_llm = get_llm_with_retry()
```

```python
print("--- Starting Blog Generation Crew ---")

result = blog_crew.kickoff(
    inputs={"topic": "latest developments in agentic AI"}
)

print("--- Blog Generation Complete ---")
print(result)
```
